### get the subject data, containing the image name and the area from where the data is coming from 

```python
len(subject_data)
```

```python
image_names_list = []
areas_list = []
annotations_list = []
retired_n_list = []
retired_bool_list = []
for i in range(len(subject_data)): 
    
    meta = json.loads(subject_data[i])
    
    retired = meta[list(meta.keys())[0]]['retired']
    
    if retired is not None: 
        retired_bool_list.append(1)
        retired_n_list.append(meta[list(meta.keys())[0]]['retired']['classifications_count'])
    else:
        retired_bool_list.append(0)
        retired_n_list.append(0)
    
    annotation = json.loads(annotations[i])
    
    if (annotation[0]['value'].upper() == 'YES'): 
        entered_data = []
        for j in range(1, len(annotation)): 
            entered_data.append(annotation[j]['value'])
        annotations_list.append(entered_data)
    else: 
        annotations_list.append([-9999.9])
        
    d = meta[list(meta.keys())[0]]
    
    image_name = d[list(d.keys())[1]]
    
    image_names_list.append(image_name)
    
    area = image_name.split('_')[-1].split('.')[0]
    
    areas_list.append(area)
```

```python
retired_bool_list_df = pd.DataFrame(retired_bool_list, columns=['retired'])
```

```python
retired_n_list_df = pd.DataFrame(retired_n_list, columns=['classifications_count'])
```

```python
areas_df = pd.DataFrame(areas_list, columns=['area'])
```

```python
image_names_df = pd.DataFrame(image_names_list, columns=['image_name'])
```

```python
l_annotations = []
for ann in annotations_list: 
    l_annotations.append(len(ann))
    
```

```python
def extend_list(l, n, pad='-9999.9'):
    if len(l) >= n:
        return l[:n]
    return l + ([pad] * (n - len(l)))
```

```python
annotations_list_padded = [] 
for l in annotations_list:
    annotations_list_padded.append(extend_list(l, 3))
    
```

```python
annotations_list_padded_df = pd.DataFrame(annotations_list_padded)
```

```python
annotations_list_padded_df
```

```python
all_concat = pd.concat([image_names_df, areas_df, retired_bool_list_df, retired_n_list_df, annotations_list_padded_df],axis=1)
```
